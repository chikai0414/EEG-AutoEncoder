```python
import pandas as pd
import numpy as np
import os
import sys, csv
```

```python
def save2csv(data, dir_path, name):
    if (os.path.isdir(dir_path)==False):
        try:
            os.makedirs(dir_path)
        except OSError:
            print ("Creation of the directory %s failed" % dir_path)
        else:
            print ("Successfully created the directory %s" % dir_path)
    
    path = dir_path+'/'+name+'.csv'
    print(path)
    data.to_csv(path, index=False)
```

```python
healthy_sample_rate = 512
illness_sample_rate = 250
healthy_path = './healthy'
illness_path = './illness'
save_path = './time_domain/'
shift = 200 #前200秒不穩定
```

```python
for dirPath, _, fileNames in os.walk(healthy_path):
    for f in fileNames:
        ds = pd.read_csv(os.path.join(dirPath, f))
        ds = ds[shift*healthy_sample_rate:] #remove 前200秒data
        length = len(ds)
        for i in range(0, length, healthy_sample_rate*4):
            start = i
            end = i+healthy_sample_rate*4
            if end > length:
                break
            mydata = ds[start:end]
            save2csv(mydata, save_path+'heatlhty/'+f.replace('.csv','')+'/', str(int(i/healthy_sample_rate))+'-'+str(int(end/healthy_sample_rate)))
```

```python
for dirPath, _, fileNames in os.walk(illness_path):
    for f in fileNames:
        ds = pd.read_csv(os.path.join(dirPath, f))
        ds = ds[shift*illness_sample_rate:] #remove 前200秒data
        length = len(ds)
        for i in range(0, length, illness_sample_rate*4):
            start = i
            end = i+illness_sample_rate*4
            if end > length:
                break
            mydata = ds[start:end]
            save2csv(mydata, save_path+'illness/'+f.replace('.csv',''), str(int(i/illness_sample_rate))+'-'+str(int(end/illness_sample_rate)))
```
